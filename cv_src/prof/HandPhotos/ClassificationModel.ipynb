Rock Paper Scissors Classification Creation Process, and what was learned.

Intro:

Our goal within this project is to play a game of "Rock Paper Scissors Minus One", and develope a system that will advise us on how to win the game. This system must be able to understand the state of our game in realtime, and provide us with the best possible move to win against our opponent as fast as possible. A system like this requires 3 main components to work together. The 3 components are as follows: Image Capture & Preperation, Image Classification, and Game Theory & Advice Output. The clear goal of this component of the project is image hand gesture classification. A tool must be developed that will accept an image of a hand and output a prediction of rock, paper, or scissors.
